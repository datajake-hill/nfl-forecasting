```python
f = open('pff.txt', 'r')
raw = f.read()
raw
```

```python
teams = ['arizona-cardinals', 'atlanta-falcons','baltimore-ravens','buffalo-bills','carolina-panthers','chicago-bears','cincinnati-bengals','cleveland-browns','dallas-cowboys','denver-broncos','detroit-lions','green-bay-packers','houston-texans','indianapolis-colts','jacksonville-jaguars','kansas-city-chiefs','los-angeles-chargers','los-angeles-rams','miami-dolphins','minnesota-vikings','new-england-patriots','new-orleans-saints','new-york-giants','new-york-jets','oakland-raiders','philadelphia-eagles','pittsburgh-steelers','san-francisco-49ers','seattle-seahawks','tampa-bay-buccaneers','tennessee-titans','washington-redskins']
years = ['2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']
# years = ['2019']
```

```python
l = raw.strip('][').replace('\'',"").split(", ")

gamedata = []
games = []
lastgame = False
year = 0
currentteam = 0

for item in l:
    item = item.split('\\n')
    if item[0] == '':
        continue
    if len(item) < 5:
        if lastgame == True:
            lastdata = []
            lastgame = False
        if item[0] in teams:
            currentteam += 1
            continue
        if item[0] in years:
            year += 1
            if year == len(years):
                year = 0
            continue
        elif len(item) == 1:
            item.append('H')
            item.append(years[year])
        else:
            item[1] = 'A'
            item.append(years[year])
        item.append(teams[currentteam])
        gamedata.append(item)
    else:
        lastgame = True
        games.append(item)
    print(item)
    
```

```python
from pandas import Series, DataFrame
import pandas as pd

headers = ['Opposing Team', 'Date', 'Time', 'W/L','PF','PA','OVR','OFF','PASS','PBLK','RECV','RUN','RBLK','DEF','RDEF','TACK','PRSH','COV','SPEC','DROP']
df = pd.DataFrame(games, columns=headers)
df = df.drop(columns=['DROP'])
metadata = pd.DataFrame(gamedata, columns=['Week','H/A','Season','Team'])
df = df.merge(metadata,left_index=True,right_index=True)
```

```python
df['Team'] = df['Team'].str.replace('-',' ').str.title()
df['Opposing Team'] = df['Opposing Team'].str.replace('-',' ').str.title()
```

```python
df = df[df['W/L'] != '-']
df
```

```python
df.head()
```

```python
all_games = df.merge(df, left_on=['Team','Opposing Team','Season','Week','Date','Time','PF'], 
         right_on=['Opposing Team','Team','Season','Week','Date','Time','PA'], 
         suffixes=['','_opp']).drop(columns=['Opposing Team_opp',
                                             'W/L_opp',
                                             'PF_opp','PA_opp',
                                             'H/A_opp','Team_opp'])
```

```python
all_games = all_games[['Team', 'Opposing Team', 'Season', 'Week', 'Date', 'Time', 'H/A', 'W/L', 'PF', 'PA', 
           'OVR', 'OFF', 'PASS', 'PBLK', 'RECV', 'RUN', 'RBLK', 'DEF', 'RDEF', 'TACK', 'PRSH',
           'COV', 'SPEC', 
           'OVR_opp', 'OFF_opp',
           'PASS_opp', 'PBLK_opp', 'RECV_opp', 'RUN_opp', 'RBLK_opp', 'DEF_opp',
           'RDEF_opp', 'TACK_opp', 'PRSH_opp', 'COV_opp', 'SPEC_opp']]
```

```python
# https://stackoverflow.com/questions/34279378/python-pandas-apply-function-with-two-arguments-to-columns
def create_true_date(s1,s2):
    x = str(s1)+'/'+str(s2)
    return x

all_games['Date'] = all_games.apply(lambda x: create_true_date(x['Date'], x['Season']), axis=1)
```
