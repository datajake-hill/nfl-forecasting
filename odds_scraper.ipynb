```python
df['Home'] = df['Home'].map(my_names)
df['Away'] = df['Away'].map(my_names)
```

```python
def point5(s):
    return s.replace('½','.5').replace('+','')

df['Home_Spread'] = df['Home_Spread'].apply(point5).astype(float)
df['Away_Spread'] = df['Away_Spread'].apply(point5).astype(float)
df['Home_Payout'] = df['Home_Payout'].apply(point5).astype(float)
df['Away_Payout'] = df['Away_Payout'].apply(point5).astype(float)
```

```python
df = df.dropna()
```
