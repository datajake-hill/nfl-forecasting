Run the code block below to create the dicitonary of correlation matrices by team and season.
Under the code there are two examples of how to use 

```python
# https://stackoverflow.com/questions/29432629/plot-correlation-matrix-using-pandas

# imports
import numpy as np
import pandas as pd
from datetime import datetime
from matplotlib import pyplot as plt

pd.set_option('display.max_columns', None)

import warnings
warnings.filterwarnings('ignore')

# begin script
pff = pd.read_csv('pff_pi.csv', index_col=0)
# pff.head(2)

teams = pff['Team'].unique()
seasons = pff['Season'].unique()

master_dict = {}
for i in teams:
    for j in seasons:
        stats = pff[(pff['Team']==i)&(pff['Season']==j)].iloc[:,8:]
        corr = stats.corr()
        master_dict[(i,j)]=corr
```

```python
# how to access a pff correlation matrix
master_dict[('Dallas Cowboys',2018)]
```

```python
# how to access an exact correlation coefficient between t
master_dict[('Dallas Cowboys',2018)].loc['OFF','DEF_opp']
```
